```python
# BITIRME PROJESI DENEME CALISMALARI-01
# DATASET ISLEMLERI

from tika import parser # pip install tika
from string import ascii_lowercase
import csv

# Burada başka kanunun odt haline çevrilmiş hali kullanılarak veri seti büyültme işlemi yapılmıştır.
raw = parser.from_file('medenwiKanun.odt')
print(raw['content'])
```

```python
metin = raw['content']
```

```python
## Gereksiz ve kelimelerden oluşmayan her şey temizlenir.
for i in range(1, 10000):
    metin = metin.replace('Madde {}-'.format(i), '')
    metin = metin.replace('{}.'.format(i), '')
    metin = metin.replace('({})'.format(i), '')
    
for c in ascii_lowercase:
    metin = metin.replace('({})'.format(c), '')
    
metin = metin.replace('\n','')
metin = metin.replace('-','')
metin = metin.replace('_','')
metin = metin.replace('.',' .')
metin = metin.replace(',',' ,')
metin = metin.replace(';',' ;')
metin = metin.replace(':',' :')
metin = metin.replace('!',' !')
metin = metin.replace('?',' ?')
metin = metin.replace('-',' - ')

metin
```

```python
print(metin)
```

```python
metin_liste = metin.split()
print(len(metin_liste))
```

```python
## Yeni bir dosya oluşturularak, dosyaya kelimeler basılır, öntanımlı olarak herbirinin etiketi 'O' yazılır.
file = open('medeniKanunv1.txt','w')
file.write("SENTENCE #\tWORD\tTAG\n")

yeni_cumle_mi = True
sayac = 1133                                                 

for i in metin_liste:    
    if(yeni_cumle_mi == True):
        file.write('Sentence : {}\t{}\tO\n'.format(sayac,i))
        sayac = sayac + 1
        yeni_cumle_mi = False
    
    else:
        file.write(' \t{}\tO\n'.format(i))
    
    if(i == '.'):
        yeni_cumle_mi = True
    
    
    

file.close()
    
```
